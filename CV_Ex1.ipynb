```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import cv2

print("opencv version: ",cv2.__version__)
```

### 1. Loading the image

(A) Load the `smarties.jpeg` and display it (correctly).

(B) Crop a single **green** "smartie" and display it separately.


```python
img = cv2.imread('smarties.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)

plt.show()
```

```python
img_crop = img[150:190,53:85,:]
plt.imshow(img_crop)
plt.show()
```

### 2. HSV histograms

(A) Convert the color space into HSV

(B) Calculate and plot a histogram of the *hue* values of all the pixels in the image (values should be between 0-180). How many "colors" can you count on that histogram?

(C) Calculate and plot the same histogram, but this time only for the cropped-green-smartie. What is the range of the hue-values of the green-smartie?


```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(16,4))
plt.hist(img_hsv[:,:,0].flatten(),255,[1,256]);
plt.xlim(0,180)
plt.xticks(np.arange(0,180,10))
plt.grid()
plt.show()
```

```python
img_hsv_crop = cv2.cvtColor(img_crop, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(16,4))
plt.hist(img_hsv_crop[:,:,0].flatten(),255,[1,256]);
plt.xlim(0,180)
plt.xticks(np.arange(0,180,10))
plt.grid()
plt.show()
```

### 3. Color mask

(A) Use `cv2.inRange` to mask all the green-smarties in the image. Then use that mask to generate an image with only the green-smarties (the rest of the image should be black).

(B) Repeat that with each and every different smarties colors (one image for each color).




```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lower = np.array([35,50,50])
upper = np.array([80,255,255])
img_mask_green = cv2.inRange(img_hsv, lower, upper)
plt.imshow(img_mask_green)
img_mask_green.shape
img_masked_green = cv2.bitwise_and(img_rgb,img_rgb, mask=img_mask_green)
plt.imshow(img_masked_green)
img_masked_green.shape
plt.show()
```

### 4. Color change

Can you change the colors of all the smarties into green-smarties ?

```python
hue = img_hsv[:,:,0];
hue[img_mask_orange==255] += 35
hue[img_mask_yellow==255] += 20
hue[img_mask_blue==255] -= 45
hue[img_mask_purple==255] -= 70
hue[img_mask_pink==255] -= 85
hue[img_mask_red==255] -= 100
plt.imshow(cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB))
plt.show()
```
